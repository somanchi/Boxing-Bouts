```python
import pandas as pd
from scipy import stats
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.metrics as sk
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv('boxing-bouts.csv')
```

```python
df_tr = df
```

```python
print(df.shape)
df.head()
```

```python
df_tr = pd.get_dummies(df_tr, columns=['result','decision'])
```

```python
clmns = ['age_A', 'age_B','height_A', 'height_B','reach_A','reach_B','weight_A','weight_B','won_A','won_B','lost_A','lost_B','drawn_A','drawn_B','kos_A','kos_B','result','judge1_A','judge1_B','judge2_A','judge2_B','judge3_A','judge3_B','decision']
```

```python
dt2 =  df.drop('result',1) #feature set
dt2 = pd.get_dummies(dt2, columns=['decision'])
```

```python
kmeans = KMeans(n_clusters=3, random_state=10).fit(dt2)
labels = kmeans.labels_
```

```python
df_tr['clusters'] = labels
```

```python
clmns.extend(['clusters'])
```

```python
print(df_tr.shape)
df_tr.head()
```

```python
dt1 = pd.read_csv('boxing-bouts-test.csv')
dt2 = dt1
```

```python
print(dt1.shape)
dt1.head()
```

```python
dt1_tr= pd.get_dummies(dt1, columns=['result'])
dt2_tr =  dt2.drop('result',1) #feature set
dt2_tr= pd.get_dummies(dt2_tr, columns=['decision'])
```

```python
y_kmeans = kmeans.predict(dt2_tr)
```

```python
centroids = kmeans.cluster_centers_
```

```python
print(centroids) # From sci-kit learn
```

```python
lable = pd.factorize(dt2['result'])[0]
```

```python
k = pd.DataFrame({'Labels': lable, 'Clusters': y_kmeans})
```

```python
ct = pd.crosstab(k['Labels'],k['Clusters'])
```

```python
print(ct)
```

```python
sk.confusion_matrix(lable,y_kmeans)
```

```python
accuracy_score(lable,y_kmeans)*100
```
